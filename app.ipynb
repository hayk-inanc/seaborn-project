```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
sns.lineplot(
    data=df, 
    x="size", 
    y="total_bill",
    estimator="mean",
    errorbar=("ci", 96),
    hue="smoker"
)
plt.show()
```

```python
sns.barplot(
    data=df, 
    x="day", 
    y="total_bill",
    estimator="median",
    hue="sex",
    errorbar=("ci", 50)
)
plt.show()
```

```python
sns.scatterplot(
    data=df,
    x="total_bill",
    y="tip"
)
plt.show()
```

```python
sns.regplot(
    data=df,
    x="total_bill",
    y="tip"
)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.scatterplot(
    data=df[df["smoker"] == "Yes"],
    x="total_bill",
    y="tip",
    ax=axes[0],
)

sns.scatterplot(
    data=df[df["smoker"] == "No"],
    x="total_bill",
    y="tip",
    ax=axes[1]
)

axes[0].set_title("Курящие")
axes[1].set_title("Не курящие")


plt.show()
```

```python
corr = df.corr(numeric_only=True)
```

```python
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    center=0,
    fmt=".2f"
)
```
